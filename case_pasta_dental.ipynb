## Desafio da solução do case: Quantas pastas dentais são vendidas no Brasil?

Usei um dataset fictício, simulando que teria dados da quantidade vendidas por mês de cada estado do Brasil incluindo o Distrito federal. 
Para trabalharmos com os dados vamos precisar:
- importar o modulo do python "csv" para trabalhar nosso dataset.
- importar o modulo pandas para renderizar nossa tabela.

```python
import csv
import pandas as pd
```

Vamos ler nosso arquivo fictício "pasta_dental.csv" que contém a quantidade por mês de cada estado brasileiro e gardalo em uma variável do tipo lista chamada data.

```python
data = []
with open('pasta_dental.csv', 'r') as csv_file:
    linhas = csv_file.read()
    linhas = linhas.split('\n')
    for linha in linhas:
        linha = linha.split(';')
        data.append(linha)
        
```

Renderizando e organizando nosso dataframe com pandas 
   
- dados mais visíveis.

```python
df = pd.read_csv('pasta_dental.csv', sep=';', engine='python')

df 
```

Após obtermos nossos dados do tipo lista podemos alteralo adicionando uma nova coluna "Brasil", onde teremos a soma do mês de cada estado, obtendo o valor do mês em nível Brasil.

- Condição ira verificar se já existe a coluna de nome Brasil para sim adicionar a coluna.
- e criando outro arquivo atualizado usando o modulo csv.

```python
if not 'Brasil' in data[0]:
    data[0].append('Brasil')
    
    for i in range( 1, len(data) ):
        soma_mes = 0
        for item in range( 1, len(data[0]) -1 ):
            soma_mes += int(data[i][item])

        data[i].append(soma_mes)
        
with open('pasta_dental-2.csv', 'w') as file:
    writer = csv.writer(file)
    writer.writerows(data)
```

## Organizando nossos dados.

Atráves de funções

- A função get_datasets(). organiza os datasets onde teremos nossa label que no caso é a coluna "Brasil" e data é uma lista dos dados de cada mês da coluna "Brasil". 

- A função set_title(). irá atráves de uma condição vai mostrar um titulo, se for passado um titulo como parametro se não, não mostrara nada.

```python
def get_datasets(y, labels):
    if type(y[0]) == list:
        datasets = []
        for i in range(len(y)):
            datasets.append({
                'label': labels[i],
                'data': y[i],
                'backgroundColor': 'rgba(54, 162, 235, 0.5)',
                'borderColor': 'rgb(54, 162, 235)',
                'borderWidth': 1,
            })
        return datasets
    else:
        return [
            {
                'label': labels[0],
                'data': y,
                'backgroundColor': 'rgba(54, 162, 235, 0.5)',
                'borderColor': 'rgb(54, 162, 235)',
                'borderWidth': 1,
            }
        ]
```

```python
def set_title(title=''):
    if title != '':
        display = 'True'
    else:
        display = 'False'
    return {
        'display': display,
        'text': title
    }
```

Precisamos criar o objeto do nosso chart para a criação do gráfico.

```python
def create_chart(x, y, labels, kind='bar', title=''):
    datasets = get_datasets(y, labels)
    options = set_title(title)
    
    chart = {
        'type': kind,
        'data': {
            'labels': x,
            'datasets': datasets
        },
        'options': {
            'title': options,
            'plugins': {
                'datalabels': {
                    'anchor': 'center',
                    'align': 'center',
                    'color': '#fff',
                    'font': {
                      'weight': 'bold',
                    },
                },
            },
        }
    }
    
    return chart
```

- Importação do módulo de requests para trabalharmos com requisições HTTP de API's.
- Import do json para enviar o objeto como json para a url da API.

```python
import json
import requests as r

def create_url(chart):
    quickchart_url = 'https://quickchart.io/chart/create'
    post_data = { 'chart': chart }
    
    response = r.post(
        quickchart_url,
        json=post_data,
    )

    if (response.status_code != 200):
        print('Error:', response.text)
    else:
        chart_response = json.loads(response.text)
    url_base = chart_response['url']
    return url_base

def get_api_chart(chart):
    url_base = create_url(chart)
    resp = r.get(url_base)
    return resp.content
```

```python
def save_image(path, content):
    with open(path, 'wb') as image:
        image.write(content)
```

Import dos subpacotes Image e display dos pacotes PIL e IPython para mostrar a imagem do nosso Gráfico com as informações trabalhadas.

```python
from PIL import Image
from IPython.display import display
```

Busca a imagem no diretorio e retorna na tela.

```python
def display_image(path):
    img_pil = Image.open(path)
    display(img_pil)
```

## Extraindo os dados e gerando o gráfico.

Aqui irá extrair os dados da nossa tabela atualizada, e usando as funções antes criadas para termos uma visualização do resultado que queremos saber.

```python
y_data_1 = []
for obs in data[1:len(data)]:
    y_data_1.append(obs[28])
    
    
labels = ['Brasil']

x = []

for obs in data[1:len(data)]:
    x.append(obs[0])
    
chart = create_chart(x, [y_data_1], labels, title='Pasta dentais vendidas no Brasil por mês em milhões.')
chart_content = get_api_chart(chart)
save_image('pasta-grafico.png', chart_content)
display_image('pasta-grafico.png')
```
